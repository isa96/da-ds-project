```python
model = keras.models.load_model('/content/drive/MyDrive/Data Capstone/Model/model_mobileNet_100_rlr_86.h5')
```

```python
import os

parrenttest = "/content/drive/MyDrive/Data Capstone/Test"

predict = []
for x in os.listdir(parrenttest) :
  path = os.path.join(parrenttest,x)
  try :
    img=load_img(path, target_size=(224,224))
    image=img_to_array(img)
    prediction_image=np.array(image)
    prediction_image= np.expand_dims(image, axis=0)
    prediction=model.predict(prediction_image)
    predict.append(np.argmax(prediction))
  except:
    print("test.csv")
```

```python
test = pd.read_csv("/content/drive/MyDrive/Data Capstone/Test/test.csv")
actual = test["label"]
fname = test["fname"]
```

```python
label = []
for i in range(len(test)) :
  if fname[i].split("_")[0] == "normal" :
    label.append(1)
  elif fname[i].split("_")[0] == "insect" :
    label.append(0)
  elif fname[i].split("_")[0] == "mould" :
    label.append(0)
  elif fname[i].split("_")[0] == "other" :
    label.append(0)
test["is_normal"] = np.array(label)

label = []
for i in range(len(test)) :
  if fname[i].split("_")[0] == "normal" :
    label.append(0)
  elif fname[i].split("_")[0] == "insect" :
    label.append(1)
  elif fname[i].split("_")[0] == "mould" :
    label.append(0)
  elif fname[i].split("_")[0] == "other" :
    label.append(0)
test["is_insect"] = np.array(label)


label = []
for i in range(len(test)) :
  if fname[i].split("_")[0] == "normal" :
    label.append(0)
  elif fname[i].split("_")[0] == "insect" :
    label.append(0)
  elif fname[i].split("_")[0] == "mould" :
    label.append(1)
  elif fname[i].split("_")[0] == "other" :
    label.append(0)
test["is_mould"] = np.array(label)


label = []
for i in range(len(test)) :
  if fname[i].split("_")[0] == "normal" :
    label.append(0)
  elif fname[i].split("_")[0] == "insect" :
    label.append(0)
  elif fname[i].split("_")[0] == "mould" :
    label.append(0)
  elif fname[i].split("_")[0] == "other" :
    label.append(1)
test["is_other"] = np.array(label)
```

```python

predict_normal=[]
for x in predict :
  if x == 3 :
    predict_normal.append(1)
  else :
    predict_normal.append(0)

predict_insect=[]
for x in predict :
  if x == 0 :
    predict_insect.append(1)
  else :
    predict_insect.append(0)

predict_mould=[]
for x in predict :
  if x == 2 :
    predict_mould.append(1)
  else :
    predict_mould.append(0)

predict_other=[]
for x in predict :
  if x == 1 :
    predict_other.append(1)
  else :
    predict_other.append(0)
```

### AUC ROC

```python
from sklearn.metrics import roc_curve, RocCurveDisplay, auc

fpr, tpr, thresholds = roc_curve(test["is_normal"],predict_normal)

roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                 estimator_name='is_normal')
display.plot()

plt.show()
```

```python
from sklearn.metrics import roc_curve, RocCurveDisplay, auc

fpr, tpr, thresholds = roc_curve(test["is_insect"],predict_insect)

roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                 estimator_name='is_insect')
display.plot()

plt.show()
```

```python
from sklearn.metrics import roc_curve, RocCurveDisplay, auc

fpr, tpr, thresholds = roc_curve(test["is_mould"],predict_mould)

roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                 estimator_name='is_mould')
display.plot()

plt.show()
```

```python
from sklearn.metrics import roc_curve, RocCurveDisplay, auc

fpr, tpr, thresholds = roc_curve(test["is_other"],predict_other)

roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                 estimator_name='is_other')
display.plot()

plt.show()
```

```python
from sklearn.metrics import roc_auc_score

norm = roc_auc_score(test["is_normal"],predict_normal)
insect = roc_auc_score(test["is_insect"],predict_insect)
mould = roc_auc_score(test["is_mould"],predict_mould)
other = roc_auc_score(test["is_other"],predict_other)
```
